```python
# preprocessing function for data
import pandas as pd
def preprocess(data):
    # load data
    df=pd.read_csv(data)
    # drop duplicates
    df=df.drop_duplicates()
    # drop null values
    df=df.dropna()
    # drop columns
    df=df.drop(columns=['col1','col2'])
    # rename columns
    df=df.rename(columns={'col1':'col2'})
    # convert to datetime
    df['col1']=pd.to_datetime(df['col1'])
    # convert to numeric
    df['col2']=pd.to_numeric(df['col2'])
    # convert to category
    df['col3']=df['col3'].astype('category')
    # convert to string
    df['col4']=df['col4'].astype('str')
    # create new column
    df['col5']=df['col1']+df['col2']
    # reorder columns
    df=df[['col1','col2','col3','col4','col5']]
    # sort rows by col1
    df=df.sort_values(by='col1',ascending=False)
    # save to csv
    df.to_csv('new_data.csv')
    return df
```

```python
import pandas as pd
from scipy import stats

def perform_tests(data):
    result = {
        'mann_whitney': [],
        'ttest': [],
        'chi_square': [],
        'shapiro_wilk': []
    }
    
    death_variable = data['death']
    
    for column in data.columns:
        if column != 'death':
            variable_data = data[column]
            
            # Check if the variable is categorical (dtype int64)
            if variable_data.dtype == 'int64':
                # Perform Chi-square test
                chi2, p_value, _, _ = stats.chi2_contingency(pd.crosstab(death_variable, variable_data))
                result['chi_square'].append((column, round(p_value, 4)))
                
            else:
                # Split the data by death variable
                group0 = variable_data[death_variable == 0]
                group1 = variable_data[death_variable == 1]
                
                # Perform Shapiro-Wilk tests
                _, p_value0 = stats.shapiro(group0)
                _, p_value1 = stats.shapiro(group1)
                
                result['shapiro_wilk'].append((column, (round(p_value0, 4), round(p_value1, 4))))
                
                # Check if both samples have a normal distribution
                if p_value0 > 0.05 and p_value1 > 0.05:
                    # Perform unpaired t-test
                    _, p_value = stats.ttest_ind(group0, group1, equal_var=False)
                    result['ttest'].append((column, round(p_value, 4)))
                else:
                    # Perform Mann-Whitney U test
                    _, p_value = stats.mannwhitneyu(group0, group1)
                    result['mann_whitney'].append((column, round(p_value, 4)))
    
    return result
```

```python
import pandas as pd
import re
from textblob import TextBlob

# Sample Twitter dataset
tweets = [
    "I love this product! It's amazing.",
    "This movie was terrible. I hated it.",
    "Just had a mediocre experience at the restaurant.",
    "Excited to go on a vacation!",
    "Feeling neutral today.",
    "I can't decide if I like this or not."
]

# Create a DataFrame from the tweets
data = {'tweets': tweets}
df = pd.DataFrame(data)

# Function to clean and analyze sentiment of a tweet
def analyze_sentiment(tweet):
    cleaned_tweet = re.sub(r'[^a-zA-Z\s]', '', tweet)  # Remove special characters
    blob = TextBlob(cleaned_tweet)
    
    # Sentiment polarity ranges from -1 (negative) to 1 (positive)
    sentiment = blob.sentiment.polarity
    
    if sentiment > 0.2:
        return 'positive'
    elif sentiment < -0.2:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to each tweet
df['sentiment'] = df['tweets'].apply(analyze_sentiment)

# Count the number of each sentiment
sentiment_counts = df['sentiment'].value_counts()

print("Sentiment Analysis Results:")
print(sentiment_counts)
```

```python
# function aims to perform AB testing on data
import numpy as np
import pandas as pd
from scipy.stats import ttest_ind

def perform_ab_test(data_control, data_variant, metric='click_through_rate', alpha=0.05):
    """
    Perform A/B testing on two datasets.

    Parameters:
        data_control (array-like): Data from the control group.
        data_variant (array-like): Data from the variant/experimental group.
        metric (str): Metric to compare (e.g., 'click_through_rate', 'conversion_rate').
        alpha (float): Significance level for the hypothesis test.

    Returns:
        result (str): A string indicating the result of the A/B test.
    """
    # Calculate the metric for both groups
    metric_control = np.mean(data_control)
    metric_variant = np.mean(data_variant)
    
    # Perform a t-test to determine if the difference is statistically significant
    t_stat, p_value = ttest_ind(data_control, data_variant)
    
    # Compare p-value with alpha
    if p_value < alpha:
        if metric_control > metric_variant:
            result = "Statistically significant - Control is better"
        else:
            result = "Statistically significant - Variant is better"
    else:
        result = "Not statistically significant - No clear winner"
    
    return result

# Simulated data for demonstration
np.random.seed(42)
data_control = np.random.normal(loc=0.5, scale=0.1, size=1000)  # Control group
data_variant = np.random.normal(loc=0.55, scale=0.1, size=1000)  # Variant group

# Perform A/B test on click-through rate
ab_test_result = perform_ab_test(data_control, data_variant, metric='click_through_rate')

print("A/B Test Result:", ab_test_result)
```

In this function, the Shapiro-Wilk test is used to determine whether the data is normally distributed or not. The null hypothesis of the test is that the data is normally distributed. If the p-value is greater than the chosen significance level (alpha), you fail to reject the null hypothesis and can conclude that the data appears to be normally distributed. Otherwise, if the p-value is less than or equal to alpha, you would reject the null hypothesis and conclude that the data does not follow a normal distribution.

```python
import numpy as np
from scipy import stats

def test_normality(data, alpha=0.05):
    """
    Test for normality of data using the Shapiro-Wilk test.
    
    Parameters:
    - data: numpy array or list, the data to be tested for normality.
    - alpha: float, the significance level for the hypothesis test.
    
    Returns:
    - is_normal: bool, True if the data appears to be normally distributed, False otherwise.
    - p_value: float, the p-value from the Shapiro-Wilk test.
    """
    data = np.array(data)
    stat, p_value = stats.shapiro(data)
    
    is_normal = p_value > alpha
    
    return is_normal, p_value

# Example usage:
data = [2.3, 3.5, 1.8, 2.9, 4.0, 2.6, 3.7, 2.1, 3.8, 2.0]
is_normal, p_value = test_normality(data)

if is_normal:
    print("The data appears to be normally distributed.")
else:
    print("The data does not appear to be normally distributed.")
    
print(f"P-value: {p_value}")
```

Ingredient Class:

The Ingredient class is used to create objects that represent individual ingredients used in animal feeds.
It has an __init__ method (constructor) that initializes an Ingredient object with four attributes: name, protein, fat, and fiber. These attributes represent the name of the ingredient and its nutritional content in terms of protein, fat, and fiber.
It also has three methods, get_protein, get_fat, and get_fiber, which are used to retrieve the protein, fat, and fiber content of the ingredient, respectively.
Feed Class:

The Feed class is used to create objects that represent animal feeds, which are composed of a list of ingredients.
It has an __init__ method that takes a list of Ingredient objects as its parameter and initializes a Feed object with this list of ingredients.
It has three methods, get_protein, get_fat, and get_fiber, which are used to calculate the total protein, fat, and fiber content of the feed by iterating through the list of ingredients and summing up their respective nutritional values.
Ingredient Instances:

The code creates three instances of the Ingredient class, representing different ingredients used in animal feeds: "Corn," "Soybean," and "Fishmeal." 

Each ingredient has its own protein, fat, and fiber content.

Ingredient Nutritional Data:

"Corn" has 8 grams of protein, 4 grams of fat, and 2 grams of fiber.
"Soybean" has 6 grams of protein, 6 grams of fat, and 6 grams of fiber.
"Fishmeal" has 12 grams of protein, 12 grams of fat, and 0 grams of fiber.

```python
# create a simple feed formulation program for chicken
class Ingredient:
    def __init__(self, name, protein, fat, fiber):
        self.name = name
        self.protein = protein
        self.fat = fat
        self.fiber = fiber
    def get_protein(self):
        return self.protein
    def get_fat(self):
        return self.fat
    def get_fiber(self):
        return self.fiber
    
class Feed:
    def __init__(self, ingredients):
        self.ingredients = ingredients
    def get_protein(self):
        total = 0
        for ingredient in self.ingredients:
            total += ingredient.get_protein()
        return total
    def get_fat(self):
        total = 0
        for ingredient in self.ingredients:
            total += ingredient.get_fat()
        return total
    def get_fiber(self):
        total = 0
        for ingredient in self.ingredients:
            total += ingredient.get_fiber()
        return total
    
ingredients = [
    Ingredient("Corn", 8, 4, 2),
    Ingredient("Soybean", 6, 6, 6),
    Ingredient("Fishmeal", 12, 12, 0)
]


feed = Feed(ingredients)
print("Protein:", feed.get_protein())
print("Fat:", feed.get_fat())
print("Fiber:", feed.get_fiber())


```

```python
# create budget class
# Define a class called 'Budget' for managing financial transactions
class Budget:
    # Constructor method to initialize a budget with a name and an amount
    def __init__(self, name, amount):
        self.name = name  # Set the budget's name
        self.amount = amount  # Set the initial budget amount

    # Method to deposit money into the budget
    def deposit(self, amount):
        self.amount += amount  # Increase the budget's amount by the deposited amount

    # Method to withdraw money from the budget
    def withdraw(self, amount):
        self.amount -= amount  # Decrease the budget's amount by the withdrawn amount

    # Method to check the current balance of the budget
    def check_balance(self):
        return self.amount  # Return the current amount in the budget

    # Method to transfer money from this budget to another budget
    def transfer(self, budget, amount):
        self.amount -= amount  # Decrease the amount in this budget
        budget.amount += amount  # Increase the amount in the target budget

# Create two budget objects with initial names and amounts
food_budget = Budget("Food", 1000)
rent_budget = Budget("Rent", 2000)

# Perform financial transactions
food_budget.deposit(500)  # Deposit 500 units into the food budget
food_budget.withdraw(200)  # Withdraw 200 units from the food budget

# Print the balances of the budgets after the transactions
print(food_budget.check_balance())
food_budget.transfer(rent_budget, 100)  # Transfer 100 units from food budget to rent budget
print(food_budget.check_balance())  # Print the updated balance of the food budget
print(rent_budget.check_balance())  # Print the updated balance of the rent budget



```

```python
# create a class for a simple bank account
# Define a class called 'BankAccount' for managing a bank account
class BankAccount:
    # Constructor method to initialize a bank account with a name and an amount
    def __init__(self, name, amount):
        self.name = name  # Set the account's name
        self.amount = amount  # Set the initial account amount

    # Method to deposit money into the account
    def deposit(self, amount):
        self.amount += amount  # Increase the account's amount by the deposited amount

    # Method to withdraw money from the account
    def withdraw(self, amount):
        self.amount -= amount  # Decrease the account's amount by the withdrawn amount

    # Method to check the current balance of the account
    def check_balance(self):
        return self.amount  # Return the current amount in the account
    
# Define a class called 'SavingsAccount' that inherits from the 'BankAccount' class
class SavingsAccount(BankAccount):
    # Constructor method to initialize a savings account with a name and an amount
    def __init__(self, name, amount):
        super().__init__(name, amount)  # Call the constructor of the parent class
        self.interest_rate = 0.01  # Set the interest rate for this account

    # Method to add interest to the account
    def add_interest(self):
        interest = self.amount * self.interest_rate  # Calculate the interest amount
        self.amount += interest  # Add the interest to the account
    
# Define a class called 'CurrentAccount' that inherits from the 'BankAccount' class
class CurrentAccount(BankAccount):
    # Constructor method to initialize a current account with a name and an amount
    def __init__(self, name, amount):
        super().__init__(name, amount)  # Call the constructor of the parent class
        self.overdraft_limit = 1000  # Set the overdraft limit for this account

    # Method to check the current balance of the account
    def check_balance(self):
        return self.amount + self.overdraft_limit  # Return the current amount in the account plus the overdraft limit

    # Method to withdraw money from the account
    def withdraw(self, amount):
        if self.amount - amount < -self.overdraft_limit:  # Check if the withdrawal will exceed the overdraft limit
            print("Withdrawal failed: overdraft limit exceeded")
        else:
            self.amount -= amount  # Decrease the account's amount by the withdrawn amount
        
# Create a savings account with initial name and amount
savings_account = SavingsAccount("Savings", 1000)
savings_account.add_interest()  # Add interest to the account
print(savings_account.check_balance())  # Print the updated balance of the account

# Create a current account with initial name and amount
current_account = CurrentAccount("Current", 1000)
current_account.withdraw(2000)  # Attempt to withdraw 2000 units from the account
print(current_account.check_balance())  # Print the updated balance of the account


```

```python
# create school management system
import sqlite3

# Create or connect to an SQLite database
conn = sqlite3.connect('school_management.db')
cursor = conn.cursor()

# Create tables for students, teachers, classes, courses, attendance, grades, notifications, and parent/guardian accounts
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    student_id TEXT PRIMARY KEY,
    name TEXT,
    grade INTEGER
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS teachers (
    teacher_id TEXT PRIMARY KEY,
    name TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS classes (
    class_id TEXT PRIMARY KEY,
    name TEXT,
    course_id TEXT,
    teacher_id TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS courses (
    course_id TEXT PRIMARY KEY,
    name TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS attendance (
    student_id TEXT,
    class_id TEXT,
    date TEXT,
    status TEXT,
    PRIMARY KEY (student_id, class_id, date)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS grades (
    student_id TEXT,
    course_id TEXT,
    assignment REAL,
    exam REAL,
    PRIMARY KEY (student_id, course_id)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS notifications (
    notification_id INTEGER PRIMARY KEY AUTOINCREMENT,
    sender TEXT,
    recipient TEXT,
    message TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS parent_guardians (
    username TEXT PRIMARY KEY,
    password TEXT,
    student_ids TEXT
)
''')

# Commit the changes and close the connection
conn.commit()
conn.close()

# ... (Previous code for classes, functions, and basic data structures)

# Function to add a student to the SQLite database
def add_student_to_db(student):
    conn = sqlite3.connect('school_management.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO students VALUES (?, ?, ?)", (student.student_id, student.name, student.grade))
    conn.commit()
    conn.close()

# Function to add a teacher to the SQLite database
def add_teacher_to_db(teacher):
    conn = sqlite3.connect('school_management.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO teachers VALUES (?, ?)", (teacher.teacher_id, teacher.name))
    conn.commit()
    conn.close()

# Function to add a class to the SQLite database
def add_class_to_db(class_obj):
    conn = sqlite3.connect('school_management.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO classes VALUES (?, ?, ?, ?)",
                   (class_obj.class_id, class_obj.name, class_obj.course_id, class_obj.teacher_id))
    conn.commit()
    conn.close()

# Function to add a course to the SQLite database
def add_course_to_db(course):
    conn = sqlite3.connect('school_management.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO courses VALUES (?, ?)", (course.course_id, course.name))
    conn.commit()
    conn.close()

# Function to mark attendance for a class and store it in the SQLite database
def mark_attendance_for_class_in_db(student_id, class_id, date, status):
    conn = sqlite3.connect('school_management.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO attendance VALUES (?, ?, ?, ?)", (student_id, class_id, date, status))
    conn.commit()
    conn.close()

# Function to enter student grades and store them in the SQLite database
def enter_student_grades_in_db(student_id, course_id, assignment, exam):
    conn = sqlite3.connect('school_management.db')
    cursor = conn.cursor()
    cursor.execute("INSERT OR REPLACE INTO grades VALUES (?, ?, ?, ?)",
                   (student_id, course_id, assignment, exam))
    conn.commit()
    conn.close()

# Function to send a notification and store it in the SQLite database
def send_notification_in_db(sender, recipient, message):
    conn = sqlite3.connect('school_management.db')
    cursor = conn.cursor()
    cursor.execute("INSERT INTO notifications (sender, recipient, message) VALUES (?, ?, ?)",
                   (sender, recipient, message))
    conn.commit()
    conn.close()

  
# User data stored in a dictionary (In a production system, use a database)
user_data = {
    "admin": "adminpassword",
    "teacher1": "teacherpassword",
    "student1": "studentpassword",
}

# Function to authenticate a user
def authenticate_user(username, password):
    # Check if the username exists in the user_data dictionary
    if username in user_data:
        # Check if the entered password matches the stored password
        if user_data[username] == password:
            return True
    return False

# Main login loop
while True:
    print("\nSchool Management System - Login")
    username = input("Enter your username: ")
    password = input("Enter your password: ")

    if authenticate_user(username, password):
        print(f"Welcome, {username}!")
        break
    else:
        print("Authentication failed. Please try again.")

# Define classes for Student and Teacher
class Student:
    def __init__(self, student_id, name, grade):
        self.student_id = student_id
        self.name = name
        self.grade = grade

class Teacher:
    def __init__(self, teacher_id, name, subject):
        self.teacher_id = teacher_id
        self.name = name
        self.subject = subject

class Class:
    def __init__(self, class_id, name, teacher_id):
        self.class_id = class_id
        self.name = name
        self.teacher_id = teacher_id
        self.students = []

    def enroll_student(self, student):
        self.students.append(student)

class Course:
    def __init__(self, course_id, name):
        self.course_id = course_id
        self.name = name
        self.classes = []

    def add_class(self, new_class):
        self.classes.append(new_class)

# Define a class for Grade
class Grade:
    def __init__(self, student_id, assignment, exam):
        self.student_id = student_id
        self.assignment = assignment
        self.exam = exam
# Define a class for Timetable
class Timetable:
    def __init__(self):
        self.schedule = {}  # Dictionary to store the schedule: {day: {time_slot: class_id}}

    def add_class_to_schedule(self, day, time_slot, class_id):
        if day not in self.schedule:
            self.schedule[day] = {}
        self.schedule[day][time_slot] = class_id


# Define a class for Attendance
class Attendance:
    def __init__(self, date):
        self.date = date
        self.attendance_record = {}

    def mark_attendance(self, student_id, status):
        self.attendance_record[student_id] = status
# Define a class for Parent or Guardian
class ParentGuardian:
    def __init__(self, username, password, student_ids):
        self.username = username
        self.password = password
        self.student_ids = student_ids  # List of student IDs linked to the parent/guardian
# Define a class for Notification
class Notification:
    def __init__(self, sender, recipient, message):
        self.sender = sender
        self.recipient = recipient
        self.message = message

# Create empty lists to store students and teachers
students = []
teachers = []
courses = []
classes = []
parent_guardians = []
notifications = []

# Function to add a new student
def add_student():
    student_id = input("Enter student ID: ")
    name = input("Enter student name: ")
    grade = input("Enter student grade: ")
    student = Student(student_id, name, grade)
    students.append(student)
    print("Student added successfully!")

# Function to add a new teacher
def add_teacher():
    teacher_id = input("Enter teacher ID: ")
    name = input("Enter teacher name: ")
    subject = input("Enter teacher subject: ")
    teacher = Teacher(teacher_id, name, subject)
    teachers.append(teacher)
    print("Teacher added successfully!")

# Function to display the list of students
def display_students():
    if not students:
        print("No students in the system.")
    else:
        print("\nList of Students:")
        for student in students:
            print(f"ID: {student.student_id}, Name: {student.name}, Grade: {student.grade}")

# Function to display the list of teachers
def display_teachers():
    if not teachers:
        print("No teachers in the system.")
    else:
        print("\nList of Teachers:")
        for teacher in teachers:
            print(f"ID: {teacher.teacher_id}, Name: {teacher.name}, Subject: {teacher.subject}")

# Function to add a new class
def add_class():
    class_id = input("Enter class ID: ")
    name = input("Enter class name: ")
    teacher_id = input("Enter teacher ID for the class: ")

    # Check if the teacher exists
    teacher = find_teacher_by_id(teacher_id)
    if not teacher:
        print("Teacher not found. Please add the teacher first.")
        return
    

    new_class = Class(class_id, name, teacher_id)
    classes.append(new_class)
    print("Class added successfully!")

# Function to add a new course
def add_course():
    course_id = input("Enter course ID: ")
    name = input("Enter course name: ")
    new_course = Course(course_id, name)
    courses.append(new_course)
    print("Course added successfully!")

# Function to find a teacher by ID
def find_teacher_by_id(teacher_id):
    for teacher in teachers:
        if teacher.teacher_id == teacher_id:
            return teacher
    return None
# Function to display classes for a course
def display_classes_for_course(course_id):
    found = False
    for course in courses:
        if course.course_id == course_id:
            print(f"\nClasses for '{course.name}':")
            for class_obj in course.classes:
                teacher = find_teacher_by_id(class_obj.teacher_id)
                print(f"Class ID: {class_obj.class_id}, Name: {class_obj.name}, Teacher: {teacher.name}")
            found = True
            break
    if not found:
        print("Course not found.")
# Function to enroll a student in a class
def enroll_student_in_class():
    class_id = input("Enter class ID: ")
    student_id = input("Enter student ID to enroll: ")

    # Check if the class and student exist
    class_obj = find_class_by_id(class_id)
    student = find_student_by_id(student_id)

    if not class_obj:
        print("Class not found.")
        return
    if not student:
        print("Student not found.")
        return

    class_obj.enroll_student(student)
    print(f"{student.name} enrolled in {class_obj.name} successfully!")
# Function to assign a teacher to a course
def assign_teacher_to_course():
    course_id = input("Enter course ID: ")
    teacher_id = input("Enter teacher ID to assign: ")

    # Check if the course and teacher exist
    course = find_course_by_id(course_id)
    teacher = find_teacher_by_id(teacher_id)

    if not course:
        print("Course not found.")
        return
    if not teacher:
        print("Teacher not found.")
        return

    course_teacher = find_teacher_by_id(course.teacher_id)
    if course_teacher:
        print(f"'{course.name}' already has a teacher: {course_teacher.name}")
        return

    course.teacher_id = teacher_id
    print(f"Teacher {teacher.name} assigned to course '{course.name}' successfully!")
# Function to find a class by ID
def find_class_by_id(class_id):
    for class_obj in classes:
        if class_obj.class_id == class_id:
            return class_obj
    return None

# Function to find a course by ID
def find_course_by_id(course_id):
    for course in courses:
        if course.course_id == course_id:
            return course
    return None

# Function to find a student by ID
def find_student_by_id(student_id):
    for student in students:
        if student.student_id == student_id:
            return student
    return None

# Function to enter student grades for an assignment and exam
def enter_student_grades():
    student_id = input("Enter student ID: ")
    assignment = float(input("Enter assignment grade (0-100): "))
    exam = float(input("Enter exam grade (0-100): "))

    # Check if the student exists
    student = find_student_by_id(student_id)
    if not student:
        print("Student not found.")
        return

    # Create a new grade entry for the student
    grade = Grade(student_id, assignment, exam)
    student.grades.append(grade)
    print(f"Grades entered for {student.name}: Assignment - {assignment}, Exam - {exam}")

# Function to calculate and display student GPAs
def calculate_and_display_student_gpas():
    for student in students:
        if student.grades:
            total_points = 0
            total_credits = 0
            for grade in student.grades:
                total_points += (grade.assignment + grade.exam) / 10
                total_credits += 1
            gpa = total_points / total_credits
            print(f"{student.name}'s GPA: {gpa:.2f}")
        else:
            print(f"{student.name} has no recorded grades.")
# Function to calculate and display class averages
def calculate_and_display_class_averages(class_id):
    class_obj = find_class_by_id(class_id)
    if not class_obj:
        print("Class not found.")
        return

    if class_obj.students:
        total_points = 0
        total_credits = 0
        for student in class_obj.students:
            if student.grades:
                for grade in student.grades:
                    total_points += (grade.assignment + grade.exam) / 10
                    total_credits += 1
        if total_credits > 0:
            class_average = total_points / total_credits
            print(f"Class {class_obj.name} Average: {class_average:.2f}")
        else:
            print(f"No grades recorded for class {class_obj.name}.")
    else:
        print(f"No students enrolled in class {class_obj.name}.")

# Function to mark attendance for a class
def mark_attendance_for_class():
    class_id = input("Enter class ID: ")
    date = input("Enter date (YYYY-MM-DD): ")

    # Check if the class exists
    class_obj = find_class_by_id(class_id)
    if not class_obj:
        print("Class not found.")
        return

    # Create an attendance record for this class session
    attendance = Attendance(date)

    # Mark attendance for each student in the class
    for student in class_obj.students:
        status = input(f"Is {student.name} present? (Y/N): ").strip().lower()
        if status == "y":
            attendance.mark_attendance(student.student_id, "Present")
        else:
            attendance.mark_attendance(student.student_id, "Absent")

    # Store the attendance record in the class
    class_obj.attendance.append(attendance)
    print(f"Attendance marked for {class_obj.name} on {date}.")

# Function to display attendance for a class
def display_attendance_for_class():
    class_id = input("Enter class ID: ")
    date = input("Enter date (YYYY-MM-DD): ")

    # Check if the class exists
    class_obj = find_class_by_id(class_id)
    if not class_obj:
        print("Class not found.")
        return

    # Find the attendance record for the specified date
    attendance_record = None
    for attendance in class_obj.attendance:
        if attendance.date == date:
            attendance_record = attendance
            break

    if attendance_record:
        print(f"\nAttendance for {class_obj.name} on {date}:")
        for student_id, status in attendance_record.attendance_record.items():
            student = find_student_by_id(student_id)
            print(f"{student.name}: {status}")
    else:
        print(f"No attendance record found for {class_obj.name} on {date}.")
# Function to schedule a class
def schedule_class():
    class_id = input("Enter class ID to schedule: ")
    day = input("Enter day (e.g., Monday): ")
    time_slot = input("Enter time slot (e.g., 9:00 AM - 10:30 AM): ")

    # Check if the class exists
    class_obj = find_class_by_id(class_id)
    if not class_obj:
        print("Class not found.")
        return

    # Check if the day and time slot are valid
    if day not in ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]:
        print("Invalid day.")
        return

    timetable.add_class_to_schedule(day, time_slot, class_id)
    print(f"Class {class_obj.name} scheduled on {day} at {time_slot}.")
# Function to generate a teacher's timetable
def generate_teacher_timetable(teacher_id):
    teacher = find_teacher_by_id(teacher_id)
    if not teacher:
        print("Teacher not found.")
        return

    print(f"Timetable for {teacher.name}:")
    for day, schedule in timetable.schedule.items():
        print(f"{day}:")
        for time_slot, class_id in schedule.items():
            class_obj = find_class_by_id(class_id)
            if class_obj:
                print(f"  {time_slot} - {class_obj.name}")
# Function to generate a student's timetable
def generate_student_timetable(student_id):
    student = find_student_by_id(student_id)
    if not student:
        print("Student not found.")
        return

    print(f"Timetable for {student.name}:")
    for class_obj in student.classes:
        day, time_slot = find_class_schedule(class_obj)
        if day and time_slot:
            print(f"{day}: {time_slot} - {class_obj.name}")

# Function to find the schedule for a class
def find_class_schedule(class_obj):
    for day, schedule in timetable.schedule.items():
        for time_slot, class_id in schedule.items():
            if class_id == class_obj.class_id:
                return day, time_slot
    return None, None
# Function to create a new parent or guardian account
def create_parent_guardian_account():
    username = input("Enter a username for the parent/guardian: ")
    password = input("Enter a password for the parent/guardian: ")
    student_ids = input("Enter comma-separated student IDs linked to this account: ").split(",")

    # Check if the entered student IDs are valid
    valid_student_ids = []
    for student_id in student_ids:
        student = find_student_by_id(student_id.strip())
        if student:
            valid_student_ids.append(student_id.strip())

    if not valid_student_ids:
        print("No valid students linked to this account. Account creation failed.")
        return

    # Create the parent/guardian account
    parent_guardian = ParentGuardian(username, password, valid_student_ids)
    parent_guardians.append(parent_guardian)
    print("Parent/Guardian account created successfully!")

# Function to authenticate a user (including parents/guardians)
def authenticate_user(username, password):
    # Check if the username is in the parent/guardian accounts
    for parent_guardian in parent_guardians:
        if parent_guardian.username == username and parent_guardian.password == password:
            return "parent_guardian", parent_guardian.student_ids

    # Check if the username is in the user_data dictionary (students, teachers, admins)
    if username in user_data:
        if user_data[username] == password:
            if username == "admin":
                return "admin", None  # Admin access
            elif username.startswith("teacher"):
                return "teacher", None  # Teacher access
            elif username.startswith("student"):
                return "student", [username]  # Student access

    return None, None  # Authentication failed
# Function to allow parents/guardians to view student information
def parent_guardian_view_student_info(username, student_ids):
    if not student_ids:
        print("No students linked to this parent/guardian account.")
        return

    for student_id in student_ids:
        student = find_student_by_id(student_id)
        if student:
            print(f"Student ID: {student.student_id}, Name: {student.name}, Grade: {student.grade}")
# ... (Previous code for classes, functions, and basic data structures)

# Function to generate a student transcript report
def generate_student_transcript(student_id):
    student = find_student_by_id(student_id)
    if not student:
        print("Student not found.")
        return

    print(f"Transcript for {student.name} (ID: {student.student_id}):")
    if not student.grades:
        print("No grades recorded.")
        return

    total_points = 0
    total_credits = 0
    for grade in student.grades:
        total_points += (grade.assignment + grade.exam) / 10
        total_credits += 1
        print(f"Course: {grade.course_name}, Assignment: {grade.assignment}, Exam: {grade.exam}")

    gpa = total_points / total_credits if total_credits > 0 else 0
    print(f"Total GPA: {gpa:.2f}")

# Function to generate a teacher workload report
def generate_teacher_workload(teacher_id):
    teacher = find_teacher_by_id(teacher_id)
    if not teacher:
        print("Teacher not found.")
        return

    print(f"Workload report for {teacher.name} (ID: {teacher.teacher_id}):")
    for course in courses:
        if course.teacher_id == teacher_id:
            classes_taught = [class_obj for class_obj in classes if class_obj.course_id == course.course_id]
            print(f"Course: {course.name}, Classes Taught: {len(classes_taught)}")

# Function to generate a class schedule report
def generate_class_schedule(class_id):
    class_obj = find_class_by_id(class_id)
    if not class_obj:
        print("Class not found.")
        return

    day, time_slot = find_class_schedule(class_obj)
    if day and time_slot:
        print(f"Schedule for {class_obj.name} (ID: {class_obj.class_id}):")
        print(f"Day: {day}, Time Slot: {time_slot}")
    else:
        print(f"No schedule found for {class_obj.name}.")

# Function to send a notification
def send_notification(sender, recipient, message):
    notification = Notification(sender, recipient, message)
    notifications.append(notification)
    print(f"Notification sent from {sender} to {recipient}: {message}")

# Function to display notifications for a user
def display_notifications(username):
    user_notifications = [notification for notification in notifications if notification.recipient == username]
    if user_notifications:
        print(f"Notifications for {username}:")
        for notification in user_notifications:
            print(f"From: {notification.sender}, Message: {notification.message}")
    else:
        print(f"No notifications for {username}.")

# Main menu
while True:
    print("\nSchool Management System")
    print("1. Add Student")
    print("2. Add Teacher")
    print("3. Add Class")
    print("4. Add Course")
    print("5. Enroll Student in Class")
    print("6. Assign Teacher to Course")
    print("7. Mark Attendance for Class")
    print("8. Display Attendance for Class")
    print("9. Enter Student Grades")
    print("10. Calculate and Display Student GPAs")
    print("11. Calculate and Display Class Averages")
    print("12. Schedule Class")
    print("13. Generate Teacher Timetable")
    print("14. Generate Student Timetable")
    print("15. Create Parent/Guardian Account")
    print("16. Parent/Guardian Login")
    print("17. Generate Student Transcript")
    print("18. Generate Teacher Workload Report")
    print("19. Generate Class Schedule Report")
    print("20. Send Notification")
    print("21. View Notifications")
    print("22. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        add_student()
    elif choice == "2":
        add_teacher()
    elif choice == "3":
        add_class()
    elif choice == "4":
        add_course()
    elif choice == "5":
        enroll_student_in_class()
    elif choice == "6":
        assign_teacher_to_course()
    elif choice == "7":
        mark_attendance_for_class()
    elif choice == "8":
        display_attendance_for_class()
    elif choice == "9":
        enter_student_grades()
    elif choice == "10":
        calculate_and_display_student_gpas()
    elif choice == "11":
        class_id = input("Enter class ID to calculate and display class averages: ")
        calculate_and_display_class_averages(class_id)
    elif choice == "12":
        schedule_class()
    elif choice == "13":
        teacher_id = input("Enter teacher ID to generate timetable: ")
        generate_teacher_timetable(teacher_id)
    elif choice == "14":
        student_id = input("Enter student ID to generate timetable: ")
        generate_student_timetable(student_id)
    elif choice == "15":
        create_parent_guardian_account()
    elif choice == "16":
        username = input("Enter username: ")
        password = input("Enter password: ")
        user_role, linked_students = authenticate_user(username, password)
        if user_role == "parent_guardian":
            parent_guardian_view_student_info(username, linked_students)
        elif user_role:
            print(f"Welcome, {username} ({user_role.capitalize()})!")
        else:
            print("Authentication failed. Please try again.")
    elif choice == "17":
        student_id = input("Enter student ID to generate transcript: ")
        generate_student_transcript(student_id)
    elif choice == "18":
        teacher_id = input("Enter teacher ID to generate workload report: ")
        generate_teacher_workload(teacher_id)
    elif choice == "19":
        class_id = input("Enter class ID to generate schedule report: ")
        generate_class_schedule(class_id)
    elif choice == "20":
        sender = input("Enter sender (e.g., Admin): ")
        recipient = input("Enter recipient (e.g., Student, Teacher, Parent): ")
        message = input("Enter message: ")
        send_notification(sender, recipient, message)
    elif choice == "21":
        username = input("Enter username to view notifications: ")
        display_notifications(username)
    elif choice == "22":
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")

```
